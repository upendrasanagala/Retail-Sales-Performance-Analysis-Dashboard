```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("SuperStoreOrders.csv", encoding="latin1")
```

```python
df.groupby("region")["sales"].sum()
```

```python
df.groupby("category")["profit"].sum()
```

```python
df.groupby("region")["sales"].sum().plot(kind="bar")
plt.title("sales by region")
plt.show()

# Created region-wise sales visualization using matplotlib.
```

```python
sns.barplot(x="category", y="profit", data=df)
plt.show() 

# profit by category chart
```

```python
df.groupby("year")["sales"].sum().plot(kind="line")
plt.title("Yearly Sales Trend")
plt.show()

# year wise sales trend
```

```python
sns.scatterplot(x="discount", y="profit", data=df)
plt.title("Discount vs Profit Relationship")
plt.show()

# Observed relationship between discount levels and profitability trends.
```
